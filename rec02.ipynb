```python
#R5b

def bin_search_safe(a, x, l=0, r=None):
    if r is None:
        r = len(a) - 1
    if l > r:
        return -1
    mid = (l + r) // 2
    if a[mid] == x:
        return mid
    if a[mid] > x:
        return bin_search_safe(a, x, l, mid - 1)
    else:
        return bin_search_safe(a, x, mid + 1, r)
```

```python
#R6b
def pow_rec_linear(a, b):
    # ベース: a^0 = 1
    if b == 0:
        return 1

    return a * pow_rec_linear(a, b - 1)
```

```python
#10
def bin_search_leftmost(a, x, l=0, r=None):
    if r is None:
        r = len(a)
    if l >= r:
        return l if (0 <= l < len(a) and a[l] == x) else -1
    mid = (l + r) // 2
    if a[mid] >= x:
        # mid も候補なので右端を mid に寄せる
        return bin_search_leftmost(a, x, l, mid)
    else:
        # mid は x 未満なので左端は右側へ
        return bin_search_leftmost(a, x, mid + 1, r)
    
bin_search_leftmost([1,1,1,1,2,2,2,5], 2)
```
