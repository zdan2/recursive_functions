課題1：階乗

正の整数 n を受け取り、その階乗 n! を計算する再帰関数 factorial(n) を作成してください。

```python
def fractoral(n):
    if n==0:
        return 1
    return n*fractoral(n-1)

fractoral(4)
```

課題2：フィボナッチ数列

n 番目のフィボナッチ数を返す再帰関数 fibonacci(n) を作成してください。

```python
def fib(n,memo=None):
    if memo is None:
        memo={0:1,1:1}
    if n==1 or n==0:
        return memo[n]
    memo[n]=fib(n-1,memo)+fib(n-2,memo)
    return memo[n]

fib(8)
```

課題3：合計

数値の配列（リスト）を受け取り、その要素の合計を計算する再帰関数 sum_array(arr) を作成してください。

```python
def sum_array(nums,i=0):
    if i==len(nums):
        return 0
    return nums[i]+sum_array(nums,i+1)

sum_array([1,2,3,4,5])
```

課題4：文字列の逆順

文字列を受け取り、その文字列を逆順にした新しい文字列を返す再帰関数 reverse_string(s) を作成してください。

```python
def reverse_string(s,i=0):
    if i==len(s):
        return ''
    return reverse_string(s,i+1)+s[i]

reverse_string('ABCDEFG')
    
```

課題5：べき乗

2つの整数 base と exponent を受け取り、base の exponent 乗を計算する再帰関数 power(base, exponent) を作成してください。

```python
def power(base,exp):
    if exp==0:
        return 1
    if exp%2==0:
        return power(base*base,exp//2)
    else:
        return base*power(base*base,exp//2)

power(2,15)
```

課題6：最大公約数

ユークリッドの互除法を用いて、2つの正の整数 a と b の最大公約数を求める再帰関数 gcd(a, b) を作成してください。

```python
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)

gcd(90,150)
```

課題7：配列の検索

数値の配列（リスト）と検索したい値 target を受け取り、配列内に target が含まれていれば True、そうでなければ False を返す再帰関数 search_array(arr, target) を作成してください。

```python
def search_array(arr,target,i=0):
    if i==len(arr):
        return False
    if arr[i]==target:
        return True
    return search_array(arr,target,i+1)

search_array([1,2,3,4,5],9)
```

課題8：回文（かいぶん）

文字列を受け取り、その文字列が回文（前から読んでも後ろから読んでも同じ文字列）であるかどうかを判定する再帰関数 is_palindrome(s) を作成してください。

```python
def is_pali(s,i=0):
    if i>len(s)//2:
        return True
    if s[i]!=s[-(i+1)]:
        return False
    return is_pali(s,i+1)

is_pali('ab')
```

課題9：数字の各桁の和

正の整数 n を受け取り、その各桁の数字の和を計算する再帰関数 sum_digits(n) を作成してください。（例：sum_digits(123) は 1 + 2 + 3 = 6）

```python
def sum_digits(nums):
    if nums==0:
        return 0
    return sum_digits(nums//10)+nums%10

sum_digits(123)
```

課題10：パスカルの三角形

パスカルの三角形の特定の行と列（row, col）の値を計算する再帰関数 pascal_triangle(row, col) を作成してください。

```python
def pascal_triangle(row,col):
    if col==0 or col>row:
        return 1
    return pascal_triangle(row-1,col-1)+pascal_triangle(row-1,col)

pascal_triangle(2,3)
```
