```python
def bin_search(a, x, l=0, r=None):
    if r is None:
        r = len(a) - 1              
    if l > r:
        return -1
    mid = (l + r) // 2
    if a[mid] == x:
        return mid
    if a[mid] > x:
        return bin_search(a, x, l, mid - 1)
    else:
        return bin_search(a, x, mid + 1, r)

bin_search([1,3,4,8,10], 8)
bin_search([1,3,4], 2)
```

```python
def pow_rec(a, b):
    if b == 0:
        return 1
    return a * pow_rec(a, b - 1)

pow_rec(2,5)
pow_rec(7,0)
```

R7 arr_sum(a)

仕様: 配列要素の合計（空配列は 0）。

例: arr_sum([2,5,1]) == 8


```python
def arr_sum(arr,i=0):
    if len(arr)==i:
        return 0
    return arr[i]+arr_sum(arr,i+1)

arr_sum([])
```

R8 arr_min(a)

仕様: 最小値を返す（空配列は ValueError でOK）。

例: arr_min([5,2,7]) == 2

```python
def arr_min(a,i=0,b=float('inf')):
    if a==[]:
        raise ValueError
    if len(a)==i:
        return b
    if b>a[i]:
        b=a[i]
    return arr_min(a,i+1,b)

arr_min([])
```

R9 find_first(a, x)

仕様: 最初に見つかった添字、なければ -1。

例: find_first([3,5,5,8], 5) == 1, find_first([1,2], 9) == -1

```python
def find_first(arr,x,i=0):
    if i==len(arr):
        return -1
    if arr[i]==x:
        return i
    return find_first(arr,x,i+1)

find_first([3,5,5,8], 10)
```

R10 bin_search_leftmost(a, x)

仕様: 重複ありの昇順配列で、x の最左の位置（なければ -1）。
ヒント：a[mid] == x のときも左側をさらに探す。

例: bin_search_leftmost([1,2,2,2,5], 2) == 1

```python
def bin_search_leftmost(a,x,l=0,r=None):
    if r==None:
        r=len(a)-1
    mid=(l+r)//2
    if l>=r:
        if a[l]==x:
            return l
        else:
            return -1

    print(l,mid,r)
    if a[mid]>x:
        return bin_search_leftmost(a,x,l,mid-1)
    elif a[mid]==x:
        return bin_search_leftmost(a,x,l,r-1)
    else:
        return bin_search_leftmost(a,x,mid+1,r)
 
   
bin_search_leftmost([1,1,1,1,1,1,2,2,2,5], 2)
```
