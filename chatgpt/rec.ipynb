R1
仕様: n==0 → 1。n は 0〜12 程度でテスト。

サンプル: factorial(5) == 120

```python
def factorials(n):
    if n==0:#ベースケース
        return 1
    return n*factorials(n-1)

factorials(5)
```

R2 sum_digits(n)

仕様: 負数は絶対値で扱う。

サンプル: sum_digits(490) == 13

```python
def sum_digits(num):
    num=abs(num)
    if num==0:#Base case
        return 0
    return num%10+sum_digits(num//10)

sum_digits(-490)
    
```

R3 reverse_str(s)

仕様: スライス（s[::-1]）禁止、再帰で。

サンプル: reverse_str("recursion") == "noisrucer"

```python
def reverse_str(s):
    if s=='':#Base case
        return ''
    return s[-1]+reverse_str(s[:-1])

reverse_str("recursion")
```

R4 is_sorted(a)

仕様: 厳密昇順（a[i] < a[i+1]）。長さ0,1は True。

サンプル: is_sorted([1,2,3]) == True, is_sorted([1,1]) == False

```python
def is_sorted(a):
    if len(a)==0 or len(a)==1:#仕様
        return True
    if a[0]>=a[1]:#大きかったらだめ
        return False
    return is_sorted(a[1:])

is_sorted([1,2,1])
```

R5 bin_search(a, x)

仕様: 昇順配列 a。見つからなければ -1。

サンプル: bin_search([1,3,4,8,10], 8) == 3, bin_search([1,3,4], 2) == -1

```python
def bin_search(a, x, l = 0, r = None):
    if r is None:
        r=len(a)
    mid=(l+r)//2
    if a[mid]==x:#Base case
        return mid
    if l>=r:
        return -1
    if a[mid]>x:
        return bin_search(a,x,l,mid-1)
    else:
        return bin_search(a,x,mid+1,r)

bin_search([1,3,4,8,10], 8)
bin_search([1,3,4], 2)
```

R6 pow_rec(a, b)

仕様: 繰り返し禁止・再帰で（まずは線形版）。後日 R13 で高速化。

サンプル: pow_rec(2, 5) == 32, pow_rec(7, 0) == 1

```python
def pow_rec(a,b):
    if b==0:
        return 1
    if b%2==1:
        return a*pow_rec(a,b//2)*pow_rec(a,b//2)
    else:
        return pow_rec(a,b//2)*pow_rec(a,b//2)

pow_rec(2,5)
pow_rec(7,0)
```

R7 arr_sum(a)

仕様: 配列要素の合計（空配列は 0）。

例: arr_sum([2,5,1]) == 8


```python
def arr_sum(arr,i=0):
    if len(arr)==i:
        return 0
    return arr[i]+arr_sum(arr,i+1)

arr_sum([])
```

R8 arr_min(a)

仕様: 最小値を返す（空配列は ValueError でOK）。

例: arr_min([5,2,7]) == 2

```python
def arr_min(a,i=0,b=float('inf')):
    if a==[]:
        raise ValueError
    if len(a)==i:
        return b
    if b>a[i]:
        b=a[i]
    return arr_min(a,i+1,b)

arr_min([])
```

R9 find_first(a, x)

仕様: 最初に見つかった添字、なければ -1。

例: find_first([3,5,5,8], 5) == 1, find_first([1,2], 9) == -1

```python
def find_first(arr,x,i=0):
    if i==len(arr):
        return -1
    if arr[i]==x:
        return i
    return find_first(arr,x,i+1)

find_first([3,5,5,8], 10)
```

R10 bin_search_leftmost(a, x)

例: bin_search_leftmost([1,2,2,2,5], 2) == 1
