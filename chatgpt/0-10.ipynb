```python
def factorial(n):
    try:
        if n<0:
            raise ValueError
        if isinstance(n,int):
            if n==0 or n==1:
                return 1
            return n*factorial(n-1)
        else:
            raise ValueError
            
    except ValueError:
        print(f'負の値や整数以外は入力しないで')
factorial(1)



def factorial(n):
    try:
        if n<0:
            raise ValueError
        if isinstance(n,int):
            res=1
            for i in range(1,n+1):
                res*=i
            return res
            
    except ValueError:
        print(f'負の値や整数以外は入力しないで')
factorial(20)
```

```python
def sum_to_n(n):
    try:
        if n<0:
            raise ValueError
        if isinstance(n,int):
            if n==0:
                return 0
            return n+sum_to_n(n-1)
        else:
            raise ValueError
            
    except ValueError:
        print(f'負の値や整数以外は入力しないで')

assert sum_to_n(0) == 0
assert sum_to_n(1) == 1
assert sum_to_n(5) == 15
for n in range(0, 200):
    assert sum_to_n(n) == n*(n+1)//2
```
