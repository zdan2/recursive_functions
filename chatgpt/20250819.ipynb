```python
def check_sum(arr, target, i = 0, cur = 0, r = None):
    if r is None:
        r = []
    if cur == target:
        return r
    if cur>target or i>=len(arr):
        return
    return check_sum(arr, target , i + 1, cur + arr[i], r + [arr[i]]) or check_sum(arr, target , i + 1, cur, r)

arr=[1,3,5]

check_sum(arr,8)
    
```
