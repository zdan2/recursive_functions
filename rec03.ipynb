```python
#R11

import random
def merge_sort(a):
    if len(a)<=1:
        return a
    n=len(a)//2
    l=merge_sort(a[:n])
    r=merge_sort(a[n:])
    return merge(l,r)

def merge(left, right):
    result_sort = [] #結合結果
    left_i, right_j = 0, 0
    #print(left,right) #=> 並び替えをしようとする配列がわかる
    while(left_i < len(left)) and (right_j < len(right)):
        if left[left_i] <= right[right_j]:
            result_sort.append(left[left_i]) #ここで結合をしようとしている rightとleftを比較しながらresultに当てていく
            left_i += 1
        else:
            result_sort.append(right[right_j]) #ここで結合をしようとしている rightとleftを比較しながらresultに当てていく
            right_j += 1
    if len(left[left_i:]) != 0: #=> 並び替えをしたが余ってしまったleftの配列内の数字が0で無ければTrue(left_iで確認）
        result_sort.extend(left[left_i:])
    if len(right[right_j:]) != 0:
        result_sort.extend(right[right_j:]) #並び替えをしたが余ってしまったrightの配列内の数字が0で無ければTrue（right_jで確認)
    #print(result_sort) #=> 並び替えた配列がわかる
    return result_sort #最終的な返り値

a=[random.randint(1,100) for _ in range(100)]
print(merge_sort(a))

```

```python
#R12 count_inversion

def count_inversions(a):#count_inverssionと同じ意味の関数

    if len(a) <= 1:
        return a, 0

    # 分割
    mid = len(a) // 2
    
    # 再帰的に統治し、「リスト」と「スワップ回数」の両方を受け取る
    left_list, left_swaps = count_inversions(a[:mid])
    right_list, right_swaps = count_inversions(a[mid:])

    # 結合処理に、左右のリストと、ここまでの合計スワップ回数を渡す
    initial_swaps = left_swaps + right_swaps
    return bubble_combine(left_list, right_list, initial_swaps)

def bubble_combine(left, right, total_swaps):
    """
    左のリストの要素を1つずつ右のリストに移し、その都度ソートを維持する。
    """
    # ベースケース: 左のリストが空になったら、完成した右のリストと合計スワップ回数を返す
    if len(left) == 0:
        return right, total_swaps

    
    # 右のリストの先頭に追加
    new_right = left + right
    
    # 新しくできたリストに対して、バブルソートの1パスをかけてソート状態にし、
    # その際のスワップ回数も受け取る
    sorted_new_right, swaps_in_pass = _bubble_pass_recursive(new_right, len(left)-1, len(left)-1)
    
    # 次の結合ステップを再帰的に呼び出す
    # 今回のパスで発生したスワップ回数を合計に加算する
    return bubble_combine([], sorted_new_right, total_swaps + swaps_in_pass)

def _bubble_pass_recursive(arr, i, j):
    # ベースケース: パスの最後まで到達した
    if i < 0:
        return arr, 0

    swaps_here = 0
    # 隣接要素を比較・交換
    if arr[i] > arr[i+1]:
        arr[i], arr[i+1] = arr[i+1], arr[i]
        swaps_here = 1
        # パスの続きを再帰的に実行し、そこでの結果を受け取る
        remaining_arr, swaps_later = _bubble_pass_recursive(arr, i + 1 ,j)
    else:
        i=j-1
        remaining_arr, swaps_later = _bubble_pass_recursive(arr, i , j )
    # 今回のスワップと、それ以降のスワップを合計して返す
    return remaining_arr, swaps_here + swaps_later


my_list = [3,2,1]
sorted_list, swap_count = count_inversions(my_list)

print(f"元のリスト: {my_list}")
print(f"ソート後のリスト: {sorted_list}")
print(f"転倒数: {swap_count}")
```

```python
#R13

def pow_fast(a,b):
    if b==0:
        return 1
    nxt=pow_fast(a,b//2)
    if b%2==1:
        return a*nxt*nxt
    else:
        return nxt*nxt

pow_fast(0,0)
    
```
